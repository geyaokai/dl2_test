```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义 PReLU 激活函数
prelu = nn.PReLU()

# 创建一个从 -5 到 5 的线性空间
x = torch.linspace(-5, 5, 100,requires_grad=True)

# 计算 PReLU 激活函数的输出
y = prelu(x)

# 计算 PReLU 激活函数的导数
y.sum().backward()  # 通过反向传播计算梯度
dy_dx = x.grad.clone() # 复制梯度

# 绘制 PReLU 激活函数
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x.detach().numpy(), y.detach().numpy(), label='PReLU')
plt.title('PReLU Activation Function')
plt.grid(True)
plt.legend()

# 绘制 PReLU 激活函数的导数
plt.subplot(1, 2, 2)
plt.plot(x.detach().numpy(), dy_dx.numpy(), label="PReLU'", color='red')
plt.title('Derivative of PReLU')
plt.grid(True)
plt.legend()

plt.show()
```
